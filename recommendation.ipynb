```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load post and user data
posts = pd.read_csv("posts.csv")
users = pd.read_csv("users.csv")

# Merge hashtags and captions
posts['content'] = posts['caption'] + ' ' + posts['hashtags']

# TF-IDF on post content
tfidf = TfidfVectorizer()
post_vectors = tfidf.fit_transform(posts['content'])

# Recommend posts for each user
for index, row in users.iterrows():
    user_interest = row['interests']
    user_vec = tfidf.transform([user_interest])
    similarity = cosine_similarity(user_vec, post_vectors)
    top_post_indices = similarity.argsort()[0][-5:][::-1]
    
    print(f"\nRecommended posts for {row['name']} ({user_interest}):")
    for i in top_post_indices:
        print(f"- {posts.iloc[i]['caption']} [{posts.iloc[i]['hashtags']}]")
```
